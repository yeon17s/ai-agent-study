```python
def return_numbers():
    return [1, 2, 3]

result1 = return_numbers()

print(type(result1))  # Output: <class 'list'>
print(return_numbers) # Output: <function return_numbers at 0x...>
print(result1) # Output: [1, 2, 3]
```

```python
def yield_numbers():
    yield 1
    yield 2
    yield 3

result2 = yield_numbers()

print(type(result2))  # Output: <class 'generator'>

# next 호출
print(next(result2))  # 1 (여기서 멈춤)
print(next(result2))  # 2 (다시 시작해서 여기서 멈춤)
print(next(result2))  # 3
```

```python
from typing import Iterable

isinstance(result2, Iterable)  
```

```python
# generator를 동시에 두 개 생성하면, 서로가 다른 객체이며, 각기 따로 동작합니다. 

i1 = yield_numbers()
i2 = yield_numbers()

print(next(i1))  
# print(next(i2))

print(next(i1))
print(next(i2))

print(next(i1))
print(next(i2))
```

```python
def yield_test():
    print('yield 1 전')
    yield 1
    print('yield 1과 2 사이')
    yield 2
    print('yield 2와 3 사이')
    yield 3
    print('yield 3 후')

i3 = yield_test()

print(next(i3))
```
