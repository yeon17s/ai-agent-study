```python
from langchain.agents import create_agent
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

@tool 
def get_weather(city: str) -> str:
  """도시 이름을 읿력받아 현재 날씨 정보를 반환합니다.

  Args:
    city: 도시 이름

  Returns:
    str: 현재 도시 날씨 정보를 반환합니다.
  """
  return '{city}의 현재 날씨는 맑고 25도입니다.'

model = ChatOpenAI(
  model="gpt-5-mini", 
  temperature=0.7
)

agent = create_agent(
  model = model,
  tools = {get_weather}
)

response = agent.invoke({
  "messages": [{
    "role": "user",
    "content": "서울 날씨 알려줘."
  }]
})

print(response['messages'][-1].content)
```
