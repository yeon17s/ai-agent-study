```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate

system_msg = SystemMessagePromptTemplate.from_template(
  "너는 도움이 되는 어시스턴트야. 한국어로만 대답해."
)
user_msg = HumanMessagePromptTemplate.from_template(
  "{city}에서 유명한 음식 3가지만 알려줘."
)

chat_template = ChatPromptTemplate.from_messages([ system_msg, user_msg ])
print(chat_template)
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template_2 = ChatPromptTemplate.from_messages(
  [
    ("system", "너는 도움이 되는 어시스턴트야. 한국어로만 대답해."),
    ("human", "{city}에서 유명한 음식 3가지만 알려줘")
  ]
)
print(chat_template_2)
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template_3 = ChatPromptTemplate.from_messages(
  [
    ("system", "너는 {language} 언어 선생님 역할을 수행할 거야."),
    ("human", "안녕, 오늘 배울 내용은 뭐야?"),
    ("ai", "안녕하세요! 오늘은 문법 기초를 한 번 배워볼까요?"),
    ("human", "{user_input}")
  ]
)
print(chat_template_3)

msg_1 = chat_template_3.invoke({
  'language':'한국어',
  'user_input':'숫자 세는 방법에 대해서 설명해줘'
})
print(msg_1)

msg_2 = chat_template_3.invoke({
  'language':'영어',
  'user_input':'요일 세는 방법에 대해서 설명해줘'
})
print(msg_2)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()

model = ChatOpenAI(
  model="gpt-5-mini"
  # api_key 설정은 load_dotenv()가 처리해주므로 생략
)

chat_template_3 = ChatPromptTemplate.from_messages(
  [
    ("system", "너는 {language} 언어 선생님 역할을 수행할 거야."),
    ("human", "안녕, 오늘 배울 내용은 뭐야?"),
    ("ai", "안녕하세요! 오늘은 문법 기초를 한 번 배워볼까요?"),
    ("human", "{user_input}")
  ]
)
print(chat_template_3)

msg_1 = chat_template_3.invoke({
  'language':'한국어',
  'user_input':'숫자 세는 방법에 대해서 설명해줘'
})
print(msg_1)

result = model.invoke(msg_1)
print(result)
```
