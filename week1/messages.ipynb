```python
from langchain_core.messages import BaseMessage

msg = BaseMessage(
  type = "Human",
  content = "대전은 어떤 곳이야?"
)

print(msg)
print(msg.type)
```

```python
from langchain_core.messages import HumanMessage

msg2 = HumanMessage(
  content = "대전은 어떤 곳이야?"
)

print(msg2)
print(msg2.type)
```

```python
from langchain_core.messages import SystemMessage

msg3 = SystemMessage(
  content="너는 10년차 시니어 개발자 역할을 수행할 거야. 한국어로만 대답해줘."
)

print(msg3)
print(msg3.type)
```
