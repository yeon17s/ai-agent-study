```python
# Pure Python Function

# 도구 1 - 날씨       def weather()
# 도구 2 - 주식 계산  def stock()
# 도구 3 - 지도       def map_api()

# llm은 어떤 기준으로 우리가 입력한 문장에서 데이터를 뽑아서 출력할까?

def add(a: int, b: int) -> int:
  # docstr
  """ 
  a와 b를 합쳐주는 함수입니다. 
  : args int, int 
  : return int a와 b가 합쳐진 값
  """

  # GoogleStyleStr
  """ 두 수를 더한다.

  Args: 
      a (int): 첫 번째 숫자
      b (int): 두 번째 숫자 
  
  Returns: 
      int: 두 수의 합

  Raises:
      ValueError: 잘못된 값 입력 시 발생

  Examples:
      result = add(1, 2)
  """
  return a + b

# 도구에 대한 docstr를 잘 작성하는 것이 중요 -> llm이 도구를 인지하고 잘 사용할 수 있음
```

```python
from langchain_core.tools import tool

@tool
def get_weather(city: str) -> str:
    """도시의 현재 날씨를 반환합니다.

    Args: 
      city: 도시의 이름 
  
    Returns: 
      str: 도시의 현재 날씨 정보 
    """
    mock_data = {
        '서울': '맑음, 22도',
        '천안': '흐림, 20도',
        '대전': '맑음, 24도',
    }
    return mock_data.get(city, '날씨 정보 없음')

result = get_weather.invoke({'city': '천안'})
print(result)

print("도구 이름: ", get_weather.name)
print("도구 설명: ", get_weather.description)
```

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()

model = ChatOpenAI(
  model="gpt-5-mini",
  temperature=0
)

tools = [ get_weather ]
model_with_tools = model.bind_tools(tools)

model_with_tools.invoke("대전의 날씨를 알려줘.")
```
