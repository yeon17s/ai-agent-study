```python
class Eatable:
  def __init__(self, name: str):
    self.name = name

  def __add__(self, other: 'Eatable') -> 'Eatable':
    return self.name + other.name
  
class Salt(Eatable):
  def __init__(self):
    super().__init__('소금')

class Bread(Eatable):
  def __init__(self):
    super().__init__('빵')

salt: Salt = Salt()
bread: Bread = Bread()

result = salt + bread

print(result)
```

```python
# or ror |

class Pipe:
  def __init__(self, func):
    self.func = func

  def __ror__(self, value):
    return self.func(value)
  
double = Pipe(lambda x: x*2)
add_tem = Pipe(lambda x: x+10)

result = 5 | double | add_tem

print(result)
```
