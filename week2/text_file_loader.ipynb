```python
from langchain_community.document_loaders import TextLoader
from langchain_core.documents import Document

loader = TextLoader(
  file_path = "../libs/001/sample1.txt",
  encoding = "utf-8"
)

docs = loader.load()

doc: Document = docs[0]

print(doc.metadata)
print(len(doc.page_content))
print(doc.page_content[:35])
```

```python
from langchain_community.document_loaders import DirectoryLoader

txt_file_loader = DirectoryLoader(
  path = "../libs/001",
  glob = "**/*.txt",
  loader_cls = TextLoader,
  loader_kwargs = {
    "encoding": "utf-8"
  },
  silent_errors = True,
  show_progress = True
)

text_docs = txt_file_loader.load()

for txt_doc in text_docs:
  print(txt_doc.metadata)
  print(len(txt_doc.page_content))
  print(txt_doc.page_content[:35])
```
