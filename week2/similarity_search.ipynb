```python
from langchain_ollama.embeddings import OllamaEmbeddings

model = OllamaEmbeddings(
  model = "nomic-embed-text-v2-moe"
)
```

```python
import numpy as np

def cosine_similarity(vec_a: np.ndarray, vec_b: np.ndarray):
  dot_product = np.dot(vec_a, vec_b)
  norm_a = np.linalg.norm(vec_a)
  norm_b = np.linalg.norm(vec_b)

  return float(dot_product / (norm_a * norm_b))
```

```python
APPLE = "사과 과일"
PEAR = "배 과일"
COMPUTER = "데스크톱 컴퓨터"
MONITOR = "컴퓨터 모니터"

docs = [APPLE, PEAR, COMPUTER, MONITOR]

vectors = model.embed_documents(docs)

apple_pear_result = cosine_similarity(np.array(vectors[0]), np.array(vectors[1]))
print(f"apple_pear_result: {apple_pear_result:.2f}")

apple_computer_result = cosine_similarity(np.array(vectors[0]), np.array(vectors[2]))
print(f"apple_computer_result: {apple_computer_result:.2f}")

apple_monitor_result = cosine_similarity(np.array(vectors[0]), np.array(vectors[3]))
print(f"apple_monitor_result: {apple_monitor_result:.2f}")

monitor_computer_result = cosine_similarity(np.array(vectors[2]), np.array(vectors[3]))
print(f"monitor_computer_result: {monitor_computer_result:.2f}")

pear_apple_result = euclidean_distance(np.array(vectors[1]), np.array(vectors[0]))
print(f"pear_apple_result: {pear_apple_result:.2f}")
```
